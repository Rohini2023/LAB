Image processing operations , histogram equilizations ,thresholding, edge detection, data augmentation, morphological operations

```python
from PIL import Image, ImageEnhance, ImageFilter,ImageOps
import numpy as np
import matplotlib.pyplot as plt

# Load an image
img = Image.open("/content/sun.jpg")

# Resize image
resized_img = img.resize((200, 200))

# Rotate image
rotated_img = img.rotate(45)

# Convert to grayscale
gray_img = img.convert("L")

# Show images
plt.subplot(2, 2, 1)
plt.imshow(img )
plt.subplot(2, 2, 2)
plt.imshow(resized_img )
plt.subplot(2, 2, 3)
plt.imshow(rotated_img )
plt.subplot(2, 2, 4)
plt.imshow(gray_img )



```

```python
gray = img.convert("L")

# Apply histogram equalization
equalized = ImageOps.equalize(gray)

# Show results
plt.subplot(2, 3, 5)
plt.imshow(gray)
plt.subplot(2, 3, 6)
plt.imshow(equalized)
plt.tight_layout()
plt.show()
```

```python
threshold_value = 128

# Convert image to grayscale
gray = img.convert("L")

# Apply thresholding
binary = gray.point(lambda p: 255 if p > threshold_value else 0)

plt.imshow(binary)
```

```python
edges = img.convert("L").filter(ImageFilter.FIND_EDGES)

plt.imshow(edges)
```

```python
from PIL import Image, ImageFilter
import numpy as np

# Load and convert the image to grayscale
img = Image.open("/content/sun.jpg").convert("L")

# Apply Sobel edge detection filter
sobel_edges = img.filter(ImageFilter.FIND_EDGES)
plt.imshow(sobel_edges)
```

```python
import cv2
import numpy as np

# Load image
img = cv2.imread("/content/sun.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Prewitt filters
prewitt_x = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(img, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Combine edges
prewitt_edges = prewitt_x + prewitt_y

plt.imshow(prewitt_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load image
img = cv2.imread("/content/sun.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Canny Edge Detector
canny_edges = cv2.Canny(img, 100, 200)

plt.imshow(canny_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = Image.open("/content/sun.jpg")
flipped = img.transpose(Image.FLIP_LEFT_RIGHT)

# Rotate Image
rotated = img.rotate(30)

# Brightness Adjustment
enhancer = ImageEnhance.Brightness(img)
bright = enhancer.enhance(1.5)  # Increase brightness
plt.subplot(1,3,1)
plt.imshow(flipped)
plt.subplot(1,3,2)
plt.imshow(rotated)
plt.subplot(1,3,3)
plt.imshow(bright)
```

```python
gray = img.convert("L")
binary = gray.point(lambda p: 255 if p > 128 else 0)

# Apply dilation (Max filter)
dilated = binary.filter(ImageFilter.MaxFilter(3))

# Apply erosion (Min filter)
eroded = binary.filter(ImageFilter.MinFilter(3))

plt.subplot(1,2,1)
plt.imshow(dilated)
plt.subplot(1,2,2)
plt.imshow(eroded)
```
